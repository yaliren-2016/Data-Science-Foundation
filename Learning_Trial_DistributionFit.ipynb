# Gaussian Mixture Model (GMM)

```python
np.random.seed(2)
x=np.concatenate([np.random.normal(0, 2, 2000), np.random.normal(5, 5, 2000), np.random.normal(3,0.5,600)])

print (len(x))
print (x[0:3])

plt.hist(x, 80, normed = True)
plt.xlim(-10, 20)
```

```python
from sklearn.mixture import GaussianMixture

x = x.reshape(-1,1)
print (x)

# x = x.reshape(1,-1)
# print (x.shape)
# print (x[0:3, ])

clf = GaussianMixture (4, max_iter = 500, random_state = 3).fit(x)

# print (clf)

xpdf = np.linspace(-10, 20, 1000)

# print (len(xpdf))
# print (xpdf[999])

# xpdf_whole = xpdf.reshape(-1,1)
# print (xpdf_whole)
# log_density = clf.score(xpdf_whole)
# print (log_density)

# xpdf0= xpdf[0].reshape(1,1)
# log_density_0 = clf.score(xpdf0)
# print (log_density_0)

# xpdf1= xpdf[1].reshape(1,1)
# log_density_1 = clf.score(xpdf1)
# print (log_density_1)

# xpdf99= xpdf[99].reshape(1,1)
# log_density_99 = clf.score(xpdf99)
# print (log_density_99)
log_density = []

for i in range(1000):
    s = clf.score(xpdf[i].reshape(1,1))   # Score is the negative log likelihood
    log_density.append(s)
    
    # print (s)
#     if (abs(s + 4.457974754919938 ) <=0.00000001):
#         print (i)

density = np.exp(log_density)
print (density)

# density = density.reshape(-1,1)

# print (len(density))
# print (density[999])

plt.hist(x, 80, normed=True, alpha=0.5)
plt.plot(xpdf, density, '-r')
plt.xlim(-10, 20)
```
